```python
# Task- Create arrays for sales, price, cost
import numpy as np
sales=np.array([[50,30,20],[60,35,25],[55,32,22],[70,40,30],[65,38,28],[80,45,35]])
sp=[100,150,200]
c=[60,90,140]
prod=["A","B","C"]
mon=[1,2,3,4,5,6]
```

```python
print(sales)
```

```python
prod_sales=np.sum(sales,axis=0)
print("Total sales volume:")
for i in range(len(prod_sales)):
    print(f"Product {prod[i]} : {prod_sales[i]}")
```

```python
#  Calculate monthly revenue
rev=sales*sp
mon_rev=np.sum(rev,axis=1)
print("monthly revenue")
for i in range(len(mon_rev)):
    print(f"Month {mon[i]} : {mon_rev[i]}")
```

```python
# Calculate monthly total cost
cost=sales*c
mon_cos=np.sum(cost,axis=1)
print("monthly cost")
for i in range(len(mon_cos)):
    print(f"Month {mon[i]} : {mon_cos[i]}")
```

```python
# Calculate profit matrix
profit=rev-cost
print("profit matrix:\n",profit)
```

```python
# Find total profit per product
t_pr_p=np.sum(profit,axis=0)
print("Total profit per product:")
for i in range(len(t_pr_p)):
    print(f"product {prod[i]} : {t_pr_p[i]}")
```

```python
# Find most profitable product
print(f"Most profitable product: Product{prod[np.argmax(t_pr_p)]}")
```

```python
# Find profit per month
t_pr_m=np.sum(profit,axis=1)
print("profit per month:") 
for i in range(len(t_pr_m)):
    print(f"Month {mon[i]} : {t_pr_m[i]}")
```

```python
# Identify best month (highest profit)
print(f"Most profitable Month: Month {mon[np.argmax(t_pr_m)]}")
```

```python
# Calculate profit margin per product
mar=t_pr_p/rev_p
print("Profit Margin")
for i in range(len(rev_p)):
    print(f"Product {prod[i]} : {mar[i]}")
```
